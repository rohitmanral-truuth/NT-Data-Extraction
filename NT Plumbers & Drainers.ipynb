```python
import selenium
import time
from selenium import webdriver

from bs4 import BeautifulSoup
import pandas as pd
```

# Plumbers and Drainers Licensing Board

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://plumberslicensing.nt.gov.au/')
```

```python
driver.find_element_by_xpath('//*[@id="plumbersTable_length"]/label/select/option[4]').click()
```

```python
pages=int(driver.find_element_by_xpath('//*[@id="plumbersTable_paginate"]/span/a[6]').text)
pages
```

```python
xpaths=[]
for j in range(1,pages+1): #creating xpaths: just change span[i] to click different links
    xpaths.append('//*[@id="plumbersTable_paginate"]/span/a[' + str(j) + ']')
#print(xpaths) 
```

```python
len(xpaths)
```

```python
all_lst=[]

for k in range(0,len(xpaths)):
    
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    headers = [i.text for i in soup.findAll('th')]
    #print(headers)
    
    data = [i.text for i in soup.findAll('td')]
    #data
    all_lst=all_lst+data
    #all_lst
    
    driver.find_element_by_xpath('//*[@id="plumbersTable_next"]').click()
    
print(headers)
all_lst  
```

```python
all_lst[-6:]
```

```python
new_data=[all_lst[i:i+6] for i in range(0,len(all_lst),6)]
new_data
```

```python
len(new_data)
```

```python
df = pd.DataFrame(new_data, columns =headers)
df
```

```python
df.to_csv('NT_Plumber_Drainer.csv')
```
