# Importing essential libraries

```python
import selenium
import time
from selenium import webdriver

from bs4 import BeautifulSoup
import pandas as pd
```

# Current NT 'A' grade electrical workers

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://electricallicensing.nt.gov.au/licensed-electricians')
```

```python
driver.find_element_by_xpath('//*[@id="elecElectricians_length"]/label/select/option[4]').click()
```

```python
pages=int(driver.find_element_by_xpath('//*[@id="elecElectricians_paginate"]/span/a[6]').text)
pages
```

```python
xpaths=[]
for j in range(1,pages+1): #creating xpaths: just change span[i] to click different links
    xpaths.append('//*[@id="elecElectricians_paginate"]/span/a[' + str(j) + ']')
#print(xpaths) 
```

```python
len(xpaths)
```

```python
all_lst=[]
headers = ['Name', 'License no.', 'Expiry date']

for k in range(0,len(xpaths)):
    
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    data = [i.text for i in soup.findAll('td')]
    data=data[:-25]
    #data
    all_lst=all_lst+data
    #all_lst
    
    driver.find_element_by_xpath('//*[@id="elecElectricians_next"]').click()
    
print(headers)
all_lst    
```

```python
all_lst[-3:]
```

```python
new_data=[all_lst[i:i+3] for i in range(0,len(all_lst),3)]
new_data
```

```python
df = pd.DataFrame(new_data, columns =headers)
df
```

```python
df.to_csv('NT_electrical_workers.csv')
```

# Current NT electrical contractors

```python
driver.get('https://electricallicensing.nt.gov.au/licensed-electricians')
```

```python
driver.find_element_by_xpath('//*[@id="elecContractors_length"]/label/select/option[5]').click()
```

```python
pages=int(driver.find_element_by_xpath('//*[@id="elecContractors_paginate"]/span/a[6]').text)
pages
```

```python
xpaths=[]
for j in range(1,pages+1): #creating xpaths: just change span[i] to click different links
    xpaths.append('//*[@id="elecContractors_paginate"]/span/a[' + str(j) + ']')
#print(xpaths) 
len(xpaths)
```

```python
all_lst=[]
headers = ['Contractor', 'License no.', 'Expiry date','	Address','Phone']

for k in range(0,len(xpaths)):
    
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    data = [i.text for i in soup.findAll('td')]
    data=data[30:]
    #data
    all_lst=all_lst+data
    #all_lst
    
    driver.find_element_by_xpath('//*[@id="elecContractors_next"]').click()
    
print(headers)
all_lst  
```
