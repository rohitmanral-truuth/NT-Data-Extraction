# Loading the `Webdriver`

```python
driver = webdriver.Chrome()
```

# Getting the access of `NT Builder's` official website

```python
driver.get('http://www.ntlis.nt.gov.au/building-practitioners/')
```

# Checking number of records present for a single alphabet

```python
driver.find_element_by_xpath('//*[@id="firstName"]').send_keys('Z')
```

```python
driver.find_element_by_xpath('//*[@id="contentwide_text"]/div[2]/form/table/tbody/tr[7]/td[2]/div/input[1]').click()
```

```python
page_source = driver.page_source
soup =  BeautifulSoup(page_source, 'lxml')
    
num=driver.find_element_by_xpath('//*[@id="contentwide_text"]/p[3]').text
num
```

```python
numbers = []
for word in num.split():
    if word.isdigit():
        numbers.append(int(word))
print(numbers[0])
```

# Creating xpaths to retrieve the details

```python
xpaths=[]
for j in range(1,int(numbers[0])+1): #creating xpaths: just change span[i] to click different links
    xpaths.append('//*[@id="contentwide_text"]/table/tbody/tr[' + str(j) + ']/td[1]/a')
#print(xpaths) 
```

```python
len(xpaths)
```

```python
xpaths[0]
```

```python
range(1,int(numbers[0]))
```

```python
c=0
for k in range(0,len(xpaths)):
    c=c+1
c
    
```

# Main program to get all the records for any single alphabet 

```python
all_lst=[]
for k in range(0,len(xpaths)): #getting the details of every license in a combination
    lst=[]
    driver.find_element_by_xpath(xpaths[k]).click()
    
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
            
    headers = [i.text for i in soup.findAll('th')]
    #print(headers)
    
    data = [i.text for i in soup.findAll('td')]
    #data
    new_data = [ x.replace('\t', '').replace('\n', '') for x in data ]
    new_data[8]=new_data[8].strip()
    #print(new_data)
        
    lst.append(new_data)
    all_lst=all_lst+lst        
    #print(all_lst)
        
    driver.find_element_by_xpath('//*[@id="contentwide_text"]/div/a[2]').click()
```

# Creating the dataframe for all the records of every single alphabet 

```python
df = pd.DataFrame(all_lst, columns= headers)
df
```

```python
len(df)
```

# Exporting the dataframe as a `.csv` file

```python
df.to_csv('NT_Builders_Z.csv')
```
